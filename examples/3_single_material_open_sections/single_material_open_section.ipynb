```python
import numpy as np
import matplotlib.pyplot as plt

# Import BeamSection module from FlightBEND
from FlightBEND.beam_section import BeamSection
```

## Z-Profile
---

```python
# Import mesh: Z-profile; Element type Q9; Mesh with 640 elements
from meshes.perfil_Z_Q9_640e  import dados as data

## Unpacking mesh data
secNodes    = data['coordinates'] 
secElems    = data['elements']
secYoung    = data['materials'][:,0]
secNu       = data['materials'][:,1]
secRho      = data['materials'][:,2]


# Instancing BeamSection object
sec = BeamSection(coordinates=secNodes,connectivity=secElems,young=secYoung,nu=secNu,rho=secRho)
```

```python
# Maximum and minumum displacement values from ABAQUS
vmax =  1.545e-1
vmin = -6.092e-2

# Twist rate [rad/m]
rate = np.radians(0.05)

## Warping
fig, ax = sec.plotWarping( twistRate=rate, levels=10, showMesh=False, showCG=False, 
                          lengthUnits='mm', vmax=vmax,vmin=vmin, 
                          cbarKwargs=dict(ticks=None)
                          )

# Remove legend and axis frame 
ax.get_legend().remove()
ax.axis('off')


## Shear stresses
fig2, ax2 = sec.plotShearStresses(degree=4, twistRate=rate, mises=True,
                                  width=.035,scale=3,figsize=(8,5),
                                  lengthUnits='mm', stressUnits='MPa',
                                  cbarKwargs=dict(shrink=.8)
                                  )

# Remove axis frame 
ax2.axis('off')
```
